```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import load_dataset

# Load Titanic dataset
titanic_data = sns.load_dataset("titanic")

# Load Tips dataset
tips_data = sns.load_dataset("tips")

```

```python
titanic_data.head()
```

```python
titanic_data['sex'].value_counts().plot(kind="pie", autopct="%.2f")
plt.show()
```

```python
plt.hist(titanic_data['age'], bins=5)
plt.show()
```

```python
sns.distplot(titanic_data['age'])
plt.show()
```

```python
# Create the box plot
sns.boxplot(x='age', data=titanic_data)

# Show plot
plt.show()
```

#MULTIVARIATE ANALYSIS


```python
sns.scatterplot(x=tips_data["total_bill"], y=tips_data["tip"])
plt.show()
```

```python

sns.scatterplot(x=tips_data["total_bill"], y=tips_data["tip"], hue=tips_data["sex"])
plt.show()
```

```python
sns.scatterplot(x=tips_data["total_bill"], y=tips_data["tip"], hue=tips_data["sex"],
style=tips_data['smoker'])
plt.show()
```

```python
sns.barplot(x=titanic_data['pclass'],y= titanic_data['age'])
plt.show()
```

```python
sns.barplot(x=titanic_data['pclass'], y=titanic_data['fare'], hue = titanic_data["sex"])
plt.show()
```

```python
sns.boxplot(x=titanic_data['sex'], y=titanic_data["age"])
plt.show()
```

```python
sns.boxplot(x=titanic_data['sex'], y=titanic_data["age"], hue=titanic_data["survived"])
plt.show()
```

```python

# KDE plot for passengers who did not survive
sns.distplot(titanic_data[titanic_data['survived'] == 0]['age'], hist=False, color="blue", label="Did Not Survive")

# KDE plot for passengers who survived
sns.distplot(titanic_data[titanic_data['survived'] == 1]['age'], hist=False, color="orange", label="Survived")

# Add labels and legend
plt.xlabel('Age')
plt.ylabel('Density')
plt.title('Age Distribution of Titanic Passengers')
plt.legend()

# Show plot
plt.show()
```

```python
cross_tab = pd.crosstab(titanic_data['pclass'], titanic_data['survived'])
sns.heatmap(cross_tab)
```

```python
# Calculate the cross-tabulation
cross_tab = pd.crosstab(titanic_data['parch'], titanic_data['survived'])

# Create the cluster map
sns.clustermap(cross_tab)
```
